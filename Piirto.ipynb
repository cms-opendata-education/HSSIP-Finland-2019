```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.path import Path

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1, aspect=1)

ax.scatter([.5],[.5], c='gold', s=100000, label="face")
ax.scatter([.35, .65], [.60, .60], c='k', s=12000, label="eyes")
ax.scatter([.35, .65], [.60, .60], c='w', s=10000, label="eyes")
ax.scatter([.30, .70], [.55, .65], c='k', s=1000, label="eyes")


X = np.linspace(.3, .7,100)
Y = -1* (X-.5)**2 + 0.40

ax.plot(X, Y, c='k', linewidth=8, label="smile")


ax.set_xlim(0,1)
ax.set_ylim(0,1)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.set_xticks([])
ax.set_yticks([])
```
