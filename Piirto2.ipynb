```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.path import Path

fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(1,1,1, aspect=1)

ax.scatter([.6],[.6], c='k', s=108000, label="face")
ax.scatter([.6],[.6], c='gold', s=100000, label="face")

ax.scatter([.575, .625], [.61, .61], c='k', s=12000, label="eyes")
ax.scatter([.575, .625], [.61, .61], c='w', s=10000, label="eyes")
ax.scatter([.575, .625], [.61, .61], c='k', s=1000, label="eyes")

X = np.linspace(.575, .625,100)
Y = 10* (X-.6)**2 + 0.575
ax.plot(X, Y, c='k', linewidth=8, label="smile")

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

ax.set_xticks([])
ax.set_yticks([])

plt.show()
```

```python
fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(1,1,1)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

ax.set_xticks([])
ax.set_yticks([])

ax.set_xlim(0,1)
ax.set_ylim(0,1)



#ears
ax.scatter([.36, .64], [.60, .60], c='k', s=10000)
ax.scatter([.36, .64], [.60, .60], c='g', s=8000)
ax.scatter([.36, .64], [.60, .60], c='lime', s=2000)


#face
ax.scatter([.5],[.5], c='k', s=108000)
ax.scatter([.5],[.5], c='forestgreen', s=100000)

#eyes
ax.scatter([.425, .575], [.525, .525], c='k', s=10000)
ax.scatter([.425, .575], [.525, .525], c='w', s=8000)
ax.scatter([.425, .575], [.525, .525], c='k', s=600)


#eyebrows
A = np.linspace(.545, .6,1000)
D = np.linspace(.455, .4,1000)
B = -2* (A-.5)**2 + 0.6
E = -2* (D-.5)**2 + 0.6
ax.plot(A,B, c='k', linewidth=8)
ax.plot(D,E, c='k', linewidth=8)

#nose
ax.scatter([.50], [.5], c='k', s=17000)
ax.scatter([.50], [.5], c='lime', s=15000)
ax.scatter([.475, .525], [.5, .5], c='k', s=1400)


#mouth
O = np.linspace(.475, .525,100)
S = -3* (O-.5)**2 + 0.4
ax.plot(O, S, c='k', linewidth=5)

plt.show()
```
